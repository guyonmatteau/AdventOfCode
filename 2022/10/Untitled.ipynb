```python
f = open('inputtest').read()[:-1].split('\n')
```

```python
instructions = f
```

```python
# part 1
cycles = []
x = 1
cycle = 1
for instruction in instructions:
    if instruction.startswith('addx'):
        cycles.append(x)
        cycles.append(x)
        cycle += 2
        x += int(instruction.split(' ')[1])
    else:
        cycles.append(x)
        cycle += 1
    
```

```python
sum(x * cycles[x-1] for x in [20, 60, 100, 140, 180, 220])
```

```python
grid = 6 * [40 * ['.']]

for i, cycle in enumerate(cycles):
    sprite = [cycle - 1, cycle, cycle + 1]
    row = int(i / 40)
    col = i % 40
    print(col, sprite)
    if col in sprite:
        grid[row][col] = '#'
    else:
        grid[row][col] = '.'
```

```python
word = ["".join([str(x) for x in y]) for y in grid]
```
